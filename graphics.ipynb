```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import moyal
```

```python
def exp(x, a=1, N=1):
    return N*np.exp(-a*x) - 0.2

def moyal_dist(x, mu=0.7, sigma=0.2, N=0.1):
    return N * moyal.pdf(x, mu, sigma)+0.02

SR_cutoff = 0.435
np.random.seed(0)  # for reproducibility
rnd = 0.015*np.random.normal(size=122)
```

```python
plt.figure(figsize=[9, 7.5])

bins = np.linspace(0.02, 0.92, 51)
bin_centers = (bins[1:]+bins[:-1])*0.5
for i in range(1, 49):
    value =exp(bin_centers[i]) + moyal_dist(bin_centers[i]) + rnd[i]
    plt.plot([bins[i], bins[i+1]], [value, value], color='lightblue', alpha=0.8)
    plt.fill_between([bins[i], bins[i+1]], [0.012, 0.012], [value, value], color='lightblue', alpha=0.4)

x_bkg = np.linspace(0.025, 0.9, 1000)

#bkg
plt.plot(x_bkg, exp(x_bkg, 1), color='teal', lw=3.5, label='background')

#sig
mu_moyal = 0.7  # Location
sigma_moyal = 0.2  # Scale
N_moyal = 0.1  # Amplitude
x_signal = x_bkg

plt.plot(x_bkg, moyal_dist(x_signal, mu_moyal, sigma_moyal, N_moyal), color='coral', lw=3.5, label='signal')
SR_cutoff = 0.435
plt.axvline(SR_cutoff, ymin=0.018, ymax=0.8, color='silver', lw=3)
plt.text(SR_cutoff + 0.18, .635, "Signal Region", fontsize=16)
plt.arrow(SR_cutoff+ 0.001, .65, .1, 0, head_width = 0.05, head_length = 0.02, lw=3, color='coral')

# Plot arrows on x and y axes
plt.arrow(0.02, 0.012, 0.95, 0, head_width=0.02, head_length=0.02, fc='k', ec='k')  # Arrow on x-axis
plt.arrow(0.02, 0.012, 0, 0.9, head_width=0.02, head_length=0.02, fc='k', ec='k')  # Arrow on y-axis

plt.xticks([])
plt.yticks([])
plt.xlim(0, 1)
plt.ylim(0, 1)
plt.axis('off')
plt.legend(fontsize=14)
plt.show
plt.savefig('./non_resonant.png')
```
