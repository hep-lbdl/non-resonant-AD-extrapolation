```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import moyal
```

```python
def exp(x, a=1, N=1):
    return N*np.exp(-a*x) - 0.2

def moyal_dist(x, mu=0.7, sigma=0.2, N=0.1):
    return N * moyal.pdf(x, mu, sigma)

SR_cutoff = 0.435
np.random.seed(0)  # for reproducibility
rnd = 0.015*np.random.normal(size=122)
```

```python
plt.figure(figsize=[9, 5.5])
ax = plt.gca()

bins = np.linspace(0.0, 0.92, 51)
bin_centers = (bins[1:]+bins[:-1])*0.5
for i in range(0, 49):
    value =exp(bin_centers[i]) + moyal_dist(bin_centers[i]) + rnd[i]
    plt.plot([bins[i], bins[i+1]], [value, value], color='lightblue', alpha=0.8)
    plt.fill_between([bins[i], bins[i+1]], [0.012, 0.012], [value, value], color='lightblue', alpha=0.4)

x_bkg = np.linspace(0.0, 0.9, 1000)

#bkg
plt.plot(x_bkg, exp(x_bkg), color='teal', lw=3.5, label='Background')

#sig
mu_moyal = 0.7  # Location
sigma_moyal = 0.2  # Scale
N_moyal = 0.1  # Amplitude
x_signal = x_bkg

plt.plot(x_bkg, moyal_dist(x_signal, mu_moyal, sigma_moyal, N_moyal), color='coral', lw=4, label='Signal')
SR_cutoff = 0.435
plt.axvline(SR_cutoff, ymin=0.15, ymax=0.71, color='silver', lw=4)
plt.arrow(SR_cutoff+ 0.001, .65, .1, 0, head_width = 0.05, head_length = 0.02, lw=4, color='silver')
plt.text(SR_cutoff-0.001, .7, "$c$", fontsize=16)



plt.text(SR_cutoff + 0.1, -.1, "Signal Region", fontsize=16)
plt.text(SR_cutoff - 0.34, -.1, "Control Region", fontsize=16)


# Plot arrows for x y z axes
arrow_width = 0.01
head_width = 0.03
lead_length = 0.03
plt.arrow(0, 0, 0.95, 0, width = arrow_width, head_width=head_width, head_length=lead_length, fc='k', ec='k', zorder = 3)  # Arrow on x-axis
plt.arrow(0, 0, 0, 0.9, width = arrow_width, head_width=head_width, head_length=lead_length, fc='k', ec='k', zorder = 3)  # Arrow on y-axis
plt.arrow(0, 0, -.1, -.1, width = arrow_width, head_width=head_width, head_length=lead_length, fc='k', ec='k', zorder = 3)  # Arrow going out


plt.xticks([])
plt.yticks([])
#plt.xlim(0, 1)
#plt.ylim(0, 1)

plt.xlabel("$m$", fontsize = 20)
plt.text(-.1, 0, r"x", fontsize = 20,  rotation = 30)
plt.ylabel("$a.u.$", fontsize = 20, labelpad = -60)

# remove axis frame
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['left'].set_visible(False)


plt.legend(fontsize=14, frameon = False)
plt.tight_layout()
plt.savefig('./nonresonant_schematic.pdf')
plt.show()
```
